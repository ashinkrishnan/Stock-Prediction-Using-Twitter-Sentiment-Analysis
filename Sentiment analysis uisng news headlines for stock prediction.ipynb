## Sentiment analysis uisng news headlines for stock prediction

```python
import pandas as pd 
#Loading dataset 
df=pd.read_csv('stockSentimentAnalysis.csv',encoding = "ISO-8859-1")
df.to_csv('myfile.csv',index=False)
df_01=pd.read_csv("myfile.csv")
df_01.head(2)
```

```python
## train test split
train = df_01[df_01['Date']<'20150101']
test = df_01[df_01['Date']> '20141231']
```

```python
train.head(2)
```

```python
test.head(2)
```

```python
## preparing textdata
corpus = train.drop(['Date','Label'],axis=1)
```

```python
## applying Redex for data cleaning
corpus.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
```

```python
## lowercasing the text data
corpus_01=corpus.apply(lambda x: x.astype(str).str.lower())
```

```python
## combining all headlines rowwise
headlines = []
for row in range(0,len(corpus_01.index)):
    headlines.append(' '.join(str(x) for x in corpus_01.iloc[row,0:25]))
```

```python
##sample 
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
```

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(3,3))
traindataset=countvector.fit_transform(headlines)
```

```python
countvector.get_feature_names()[50:60]
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
# predictions = randomclassifier.predict(test_dataset)
predictions = randomclassifier.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
